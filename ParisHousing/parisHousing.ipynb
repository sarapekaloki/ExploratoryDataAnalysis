# Paris Housing exploratory data analysis

```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
paris_housing = pd.read_csv("ParisHousing.csv")
paris_housing
```

### Descriptive Statistics

```python
#General Statistics
paris_housing.describe().round()
```

```python
#Distribution of square meters
#The standard deviation and the mean for this value are really high, abnormal size of a property in such a big city
plt.figure(figsize=(8, 6))
sns.histplot(paris_housing['squareMeters'], kde=True)


# add labels and title
plt.xlabel('Square meters')
plt.ylabel('Frequency')
plt.title('Distribution of square meters of property')
plt.show()
```

```python
#Correlation between price and square meters
paris_housing.plot(x="squareMeters", y="price", style="o")
```
